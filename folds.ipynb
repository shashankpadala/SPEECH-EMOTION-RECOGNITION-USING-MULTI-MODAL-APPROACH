```
import pickle
import numpy as np

train=[]
for j in range(8,9):
  print("j th file",j)
  data=pickle.load(open('/content/drive/My Drive/data/mixed'+str(j)+'trainx.pickle','rb'))
  c=0
  for l1 in range(0,len(data),30):
    print("l",l1)
    max=0
    for k1 in range(l1,l1+30):
      c=c+1
      if(k1>=len(data)):
        break
      audio_data=data[k1]
      max1=audio_data[0].shape[1]
      print(max1)
      if(max<max1):
        max=max1
    print("Maximum is:",max)
    d=0
    for i in range(l1,l1+30):
      print("In files:",d)
      d=d+1
      if(i>=len(data)):
        break
      audio_data=data[i]
      lst=[]
      for k in range(0,38):
        col_add=[]
        for l in range(audio_data[0].shape[1],max):
          col_add.append(0)
        lst.append(col_add)
      lst.append(col_add)
      lst1=np.array(lst)
      print("lst",lst1.shape)
      aud = np.concatenate((audio_data[0], lst1), axis=1)
      print("shape after",aud.shape)
      data[i][0]=aud
    pickle.dump(data,open('/content/drive/My Drive/practice/padded_mixed/'+str(j)+'trainx.pickle','wb'))

print("DONE")
```
