```
import librosa
import numpy as np
import scipy
from scipy.signal import hann
import glob
import pickle
```

```
files = pickle.load( open( "/content/drive/My Drive/IEMOCAP_full_release/data.pickle", "rb" ) )
```

```
window=hann(1600)
count=0
for i in files:
  print(count)
  count=count+1
  id=i['id']
  path='/content/drive/My Drive/IEMOCAP_full_release/wav/'
  id=path+id+'.wav'
  signal,rate=librosa.load(id,sr=16000)
  audio=[]
  for j in range(int(np.shape(signal)[0]/800)):
    l1=signal[:1600] * window
    audio=np.append(audio,l1)
    np.roll(signal,-800)
  mfcc=librosa.feature.mfcc(audio,sr=16000,n_mfcc=13)
  delta=librosa.feature.delta(mfcc)
  delta2=librosa.feature.delta(mfcc,order=2)
  audio1=np.concatenate((mfcc,delta,delta2),axis=0)
  print(np.shape(audio1))
  path='/content/drive/My Drive/preprocessed audio data/'+i['id']+'.npy'
  np.save(path,audio1)
```
